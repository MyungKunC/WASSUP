```python
for i in range(2, 10):
  for j in range(1, 10):
    print(i*j, end=" ")
  print(' ')
```

```python
for i in range(2, 10):
  for j in range(1, 10):
    print(i*j, end=" ")
  print()
```

```python
a = [1,2,3,4]
result = []
for num in a:
  result.append(num*3)

print(result)
```

```python
a = [1,2,3,4]
result = [num * 3 for num in a]
print(result)
```

```python
a = [1,2,3,4]
result = [num * 3 for num in a if num % 2 == 0]
print(result)
```

```python
result = [x*y for x in range(2,10)
              for y in range(1,10)]
print(result)
```

```python
import time

start = time.time()
numbers = [i for i in range(1000000)]
end = time.time()
print('List Comprehension Time:', end-start)

start = time.time()
numbers = []
for i in range(1000000):
    numbers.append(i)
end = time.time()
print('For loop Time:', end-start)
```
